```python
import os
import json
import numpy as np

from collections import defaultdict

from utils.manager import UEManager
```

```python
man_dir = '/Users/ekaterinafadeeva/data/ue_mans/'
low = defaultdict(list)
high = defaultdict(list)
for subdir, _, man_files in os.walk(man_dir):
    if subdir != man_dir:
        continue
    for man_file in man_files:
        man = UEManager.load(os.path.join(man_dir, man_file))
        for key in man.estimations.keys():
            ue = np.array(man.estimations[key])
            low[key].append(np.quantile(ue, 0.01))
            high[key].append(np.quantile(ue, 0.99))
```

```python
ue_bounds = {}
for key in low.keys():
    level, ue = key
    ue_bounds[ue] = {
        'low': np.mean(low[key]),
        'high': np.mean(high[key]),
    }
json.dump(ue_bounds, open('utils/ue_bounds.json', 'w'))
```
